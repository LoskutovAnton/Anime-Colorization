```python
import glob
import os
from PIL import Image
```

```python
DATA_PATH = "data/"
COLOR_IMAGE_PATH = os.path.join(DATA_PATH, "color/")
GRAY_IMAGE_PATH = os.path.join(DATA_PATH, "gray/")
CONTOUR_IMAGE_PATH = os.path.join(DATA_PATH, "contour/")
FRAME_PREFIX = "frame_"
```

```python
def get_files(dir_name, ext):
    return glob.glob(dir_name + "*" + ext)

def check_dir(dir_name):
    if not os.path.exists(dir_name):
        os.makedirs(dir_name)
```

```python
old_videos = get_files("", ".mkv")
for video in old_videos:
    !ffmpeg -i {video} -ss 67 -to 1230 -async 1 -an -vf scale=320:240 {video}_new.mkv
```

```python
videos = get_files("", "new.mkv")
check_dir(COLOR_IMAGE_PATH)
frame_name = os.path.join(COLOR_IMAGE_PATH, FRAME_PREFIX)
for i, video in enumerate(videos):
    !ffmpeg -i {video} -start_number {1163*i} -f image2 -vf fps=1 {frame_name}%d.jpg
```

```python
check_dir(GRAY_IMAGE_PATH)
color_img = get_files(COLOR_IMAGE_PATH, ".jpg")

j = 0
for img_path in color_img:
    if (j % 1000 == 0):
        print(j)
    
    name = FRAME_PREFIX + str(j)
    j += 1
    img = Image.open(img_path).convert('LA')
    img.save(GRAY_IMAGE_PATH + name + '.png')
```
